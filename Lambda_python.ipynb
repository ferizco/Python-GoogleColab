lambda adalah fungsi anonymous atau fungsi yang tidak memiliki nama. lambda hanya bisa memuat satu expression saja. 

```python
kuadrat = (lambda x,y: x**2 + y**2)(4,4)
print (kuadrat)

(lambda x,y: x**2 + y**2)(4,4) #lambda juga bisa dieksekusi secara langsung dengan menambahkan tanda kurung
```

pada contoh diatas, ekspresi lambda dimasukkan kedalam sebuah variabel. lalu variabel tersebut di panggil dengan fungsi print
